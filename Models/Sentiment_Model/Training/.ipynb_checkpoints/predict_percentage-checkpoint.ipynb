```python
%%time
def preprocess_text(text):
    doc = nlp(text)
    lemma = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    joined = ' '.join(lemma)
    doc = nlp(joined)
    return doc.vector

def funct1(df):
    df['processed_text'] = df['text'].apply(preprocess_text)
    return df

# Convert pandas DataFrame to Dask DataFrame
dask_df = dd.from_pandas(finalized_data, npartitions=40)  # Adjust npartitions as needed

# Apply the function
#dask_df = dask_df.map_partitions(funct1)
dask_df['vector'] = dask_df['text'].map(preprocess_text, meta=('vector', np.empty((0, 300), dtype=np.float32)))


# Compute results
finalized_data = dask_df.compute()
print(finalized_data.head())
```

```python
class NewsDataset(Dataset):
    def __init__(self, X, y_short):
        self.X = X
        self.y_short = y_short

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y_short[idx]

# Create Dataset instances
train_dataset = NewsDataset(X_train, y_short_train)
test_dataset = NewsDataset(X_test, y_short_test)

# Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=1024, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=1024, shuffle=False)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define the neural network architecture with bidirectional LSTM layers
class DynamicLSTMModel(nn.Module):
    def __init__(self, layer_sizes):
        super(DynamicLSTMModel, self).__init__()
        
        self.layer_sizes = layer_sizes
        self.lstm_layers = nn.ModuleList()
        
        # Create LSTM layers
        for i in range(len(layer_sizes) - 1):
            self.lstm_layers.append(nn.LSTM(input_size=layer_sizes[i], 
                                            hidden_size=layer_sizes[i+1], 
                                            batch_first=True))
        
        # Define the fully connected layer
        self.fc = nn.Linear(layer_sizes[-1], 1)
    
    def forward(self, x):
        # Pass input through all LSTM layers
        for lstm in self.lstm_layers:
            x, _ = lstm(x)
        
        # Pass the output from the last LSTM layer through the fully connected layer
        out = self.fc(x)
        
        return out

# Initialize the model with the specified layers
layers = [300, 300, 150, 1]  # Example layer sizes
model = DynamicLSTMModel(layers)

# Ensure to set the device to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Define the loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
# Define the LSTM model as shown before
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, 
                            batch_first=True, 
                            bidirectional=True)
        
        self.fc = nn.Linear(hidden_size * 2, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers * 2, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers * 2, x.size(0), self.hidden_size).to(x.device)

        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out

# Hyperparameters
input_size = 300
hidden_size = 400
num_layers = 4
output_size = 1


learning_rate = 0.0001

model = LSTMModel(input_size, hidden_size, num_layers, output_size)
criterion = nn.SmoothL1Loss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python


device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Training loop
num_epochs = 50000
print_every = 100

alpha = 0.1

# Best model tracking
best_model = None
best_loss = float('inf')

for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    total_mse = 0
    total_variance_penalty = 0
    
    # Interrupt training if 'F2' is pressed (optional)
    if keyboard.is_pressed('f2'):
        print("-----Training interrupted------")
        break
    
    # Training loop
    for X_batch, y_short_batch in train_loader:
        X_batch = X_batch.unsqueeze(1)
        X_batch, y_short_batch = X_batch.to(device), y_short_batch.to(device)
        #print(X_batch.shape)
        optimizer.zero_grad()
        
        # Forward pass
        short_term_output = model(X_batch)
        predictions = short_term_output.squeeze()
        
        # Calculate loss
        mse = criterion(predictions, y_short_batch)
        
        # Variance of predictions and targets
        prediction_var = torch.var(predictions)
        y_short_batch_var = torch.var(y_short_batch)
        
        # Variance Penalty (difference between target variance and predicted variance)
        variance_penalty = abs(prediction_var - y_short_batch_var)# ** 2
        
        # Combined loss
        loss = mse + alpha * variance_penalty
                
        # Backward pass and optimization
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        total_mse += mse.item()
        total_variance_penalty += variance_penalty.item()
    
    # Validation loop
    model.eval()
    val_loss = 0

    with torch.no_grad():
        for X_batch, y_short_batch in test_loader:
            X_batch = X_batch.unsqueeze(1)
            X_batch, y_short_batch = X_batch.to(device), y_short_batch.to(device)

            short_term_output = model(X_batch)
            
            # Calculate validation loss
            val_loss += criterion(short_term_output.squeeze(), y_short_batch.squeeze()).item()

    val_loss /= len(test_loader)

    # Check if this is the best model
    if val_loss < best_loss:
        best_loss = val_loss
        best_model = deepcopy(model)
        print(f"New best model found at epoch {epoch+1} with validation loss: {val_loss:.4f}")

    if (epoch + 1) % print_every == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Total Loss: {total_loss/len(train_loader):.4f}, '
              f'MSE Loss: {total_mse/len(train_loader):.4f}, '
              f'Alpha Loss (scaled): {alpha *total_variance_penalty/len(train_loader):.4f} '
              f'Alpha Loss: {total_variance_penalty/len(train_loader):.4f}, '
              f'Val Loss: {val_loss:.4f}')

# At the end, best_model contains the model with the lowest validation loss.
```

```python
# Define the neural network architecture with variable layers
class VariableNN(nn.Module):
    def __init__(self, layers):
        super(VariableNN, self).__init__()
        self.layers = nn.ModuleList()
        self.num_layers = len(layers)
        
        # Create layers
        for i in range(self.num_layers - 1):
            self.layers.append(nn.Linear(layers[i], layers[i + 1]))
            if i < self.num_layers - 2:  # Apply ReLU activation except for the last layer
                self.layers.append(nn.Tanh())
    
    def forward(self, x):
        for layer in self.layers:
            x = layer(x)
        return x

# Initialize the model with a list of layers
layer = [300, 3000, 3000, 1000, 1]  # Example layer sizes
model = VariableNN(layer)

# Ensure to set the device to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Define loss function and optimizer
#criterion = nn.MSELoss()
criterion = nn.SmoothL1Loss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

class DeepImpactPredictor(nn.Module):
    def __init__(self, input_dim, hidden_dim, lstm_hidden_dim, output_dim, num_layers=1, dropout=0.2):
        super(DeepImpactPredictor, self).__init__()
        
        # LSTM layer
        self.lstm = nn.LSTM(input_dim, lstm_hidden_dim, num_layers=num_layers, batch_first=True, dropout=dropout)

        # Dropout layer
        self.dropout = nn.Dropout(dropout)
        
        # Fully connected layers
        self.fc1 = nn.Linear(lstm_hidden_dim, hidden_dim)
        self.leaky_relu = nn.LeakyReLU(negative_slope=0.01)
        self.fc_short_term = nn.Linear(hidden_dim, output_dim)
    
    def forward(self, x):
        # Pass through LSTM layer
        x, _ = self.lstm(x)
        
        # Use the output of the last LSTM cell
        x = x[:, -1, :]  # Take the output from the last time step

        x = self.dropout(x)
        
        # Pass through fully connected layers
        x = self.leaky_relu(self.fc1(x))
        short_term_output = self.fc_short_term(x)
        
        return short_term_output

# Parameters
input_dim = X.shape[1]  # Dimension of spaCy embeddings
hidden_dim = 64
lstm_hidden_dim = 64  # Dimension of LSTM hidden state
output_dim = 1  # Single output value for regression
num_layers = 2  # Number of LSTM layers
dropout = 0.1

# Initialize the model
model = DeepImpactPredictor(input_dim, hidden_dim, lstm_hidden_dim, output_dim, num_layers, dropout)
model.to(device)

# Loss function and optimizer for regression
criterion = nn.MSELoss()  # Mean Squared Error Loss for regression
optimizer = optim.Adam(model.parameters(), lr=0.001)
```
