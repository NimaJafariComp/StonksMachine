```python
%%time
def preprocess_text(text):
    doc = nlp(text)
    return [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
def text_to_vector(text):
    doc = nlp(text)
    return doc.vector
nlp = spacy.load('en_core_web_lg')

# Convert your pandas DataFrame to a Dask DataFrame
dask_df = dd.from_pandas(finalized_data, npartitions=24)  # Adjust npartitions based on your CPU

# Apply the preprocess_text function and join lemmas
dask_df['processed_text'] = dask_df['text'].map_partitions(lambda df: df.apply(preprocess_text))
# Ensure the processed_text column is space-separated
dask_df['processed_text'] = dask_df['processed_text'].map_partitions(lambda df: df.apply(lambda x: ' '.join(x)))
# Vectorize the processed text
dask_df['vector'] = dask_df['processed_text'].map_partitions(lambda df: df.apply(text_to_vector))

# Compute the results back into a pandas DataFrame
finalized_data = dask_df.compute()

# Convert vectors to PyTorch tensors
X_np = np.array(finalized_data['vector'].tolist())
X = torch.tensor(X_np, dtype=torch.float32)
```

```python

y_np = np.array(finalized_data['short_term_return'])

# Normalize y
scaler = StandardScaler()
y_normalized = y_np#scaler.fit_transform(y_np.reshape(-1, 1)).flatten()

# Convert normalized y back to tensor
y_short = torch.tensor(y_normalized, dtype=torch.float32)
```

```python
# Split the data
X_train, X_test, y_short_train, y_short_test = train_test_split(
    X, y_short, test_size=0.2, random_state=42)
```

```python
class NewsDataset(Dataset):
    def __init__(self, X, y_short):
        self.X = X
        self.y_short = y_short

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y_short[idx]

# Create Dataset instances
train_dataset = NewsDataset(X_train, y_short_train)
test_dataset = NewsDataset(X_test, y_short_test)

# Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=512, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=512, shuffle=False)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

class DeepImpactPredictor(nn.Module):
    def __init__(self, input_dim, hidden_dim, lstm_hidden_dim, output_dim, num_layers=1, dropout=0.2):
        super(DeepImpactPredictor, self).__init__()
        
        # LSTM layer
        self.lstm = nn.LSTM(input_dim, lstm_hidden_dim, num_layers=num_layers, batch_first=True, dropout=dropout)

        # Dropout layer
        self.dropout = nn.Dropout(dropout)
        
        # Fully connected layers
        self.fc1 = nn.Linear(lstm_hidden_dim, hidden_dim)
        self.leaky_relu = nn.LeakyReLU(negative_slope=0.01)
        self.fc_short_term = nn.Linear(hidden_dim, output_dim)
    
    def forward(self, x):
        # Pass through LSTM layer
        x, _ = self.lstm(x)
        
        # Use the output of the last LSTM cell
        x = x[:, -1, :]  # Take the output from the last time step

        x = self.dropout(x)
        
        # Pass through fully connected layers
        x = self.leaky_relu(self.fc1(x))
        short_term_output = self.fc_short_term(x)
        
        return short_term_output

# Parameters
input_dim = X.shape[1]  # Dimension of spaCy embeddings
hidden_dim = 64
lstm_hidden_dim = 64  # Dimension of LSTM hidden state
output_dim = 1  # Single output value for regression
num_layers = 2  # Number of LSTM layers
dropout = 0.1

# Initialize the model
model = DeepImpactPredictor(input_dim, hidden_dim, lstm_hidden_dim, output_dim, num_layers, dropout)
model.to(device)

# Loss function and optimizer for regression
criterion = nn.MSELoss()  # Mean Squared Error Loss for regression
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Define the neural network architecture with variable layers
class VariableNN(nn.Module):
    def __init__(self, layers):
        super(VariableNN, self).__init__()
        self.layers = nn.ModuleList()
        self.num_layers = len(layers)
        
        # Create layers
        for i in range(self.num_layers - 1):
            self.layers.append(nn.Linear(layers[i], layers[i + 1]))
            if i < self.num_layers - 2:  # Apply ReLU activation except for the last layer
                self.layers.append(nn.Tanh())
    
    def forward(self, x):
        for layer in self.layers:
            x = layer(x)
        return x

# Initialize the model with a list of layers
layer = [300, 3000, 3000, 1000, 1]  # Example layer sizes
model = VariableNN(layer)

# Ensure to set the device to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Define loss function and optimizer
#criterion = nn.MSELoss()
criterion = nn.SmoothL1Loss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Training loop
num_epochs = 50000
print_every = 100

# Best model tracking
best_model = None
best_loss = float('inf')

for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    
    # Interrupt training if 'F2' is pressed (optional)
    if keyboard.is_pressed('f2'):
        print("-----Training interrupted------")
        break
    
    # Training loop
    for X_batch, y_short_batch in train_loader:
        X_batch, y_short_batch = X_batch.to(device), y_short_batch.to(device)
        
        optimizer.zero_grad()
        
        # Forward pass
        short_term_output = model(X_batch)
        
        # Calculate loss
        loss_short_term = criterion(short_term_output.squeeze(), y_short_batch)
        loss = loss_short_term
        
        # Backward pass and optimization
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    # Validation loop
    model.eval()
    val_loss = 0

    with torch.no_grad():
        for X_batch, y_short_batch in test_loader:
            X_batch, y_short_batch = X_batch.to(device), y_short_batch.to(device)

            short_term_output = model(X_batch)
            
            # Calculate validation loss
            val_loss += criterion(short_term_output.squeeze(), y_short_batch).item()

    val_loss /= len(test_loader)

    # Check if this is the best model
    if val_loss < best_loss:
        best_loss = val_loss
        best_model = deepcopy(model)
        print(f"New best model found at epoch {epoch+1} with validation loss: {val_loss:.4f}")

    if (epoch + 1) % print_every == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss/len(train_loader):.4f}, '
              f'Val Loss: {val_loss:.4f}')

# At the end, best_model contains the model with the lowest validation loss.
```

```python
# Evaluation
#model_test = model
model_test = best_model
# Make sure the model is in evaluation mode
model_test.eval()

# Store predictions and actual values
predictions = []
actuals = []

with torch.no_grad():
    # Process each batch in the test set
    for X_batch, y_batch in test_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        
        # Add an additional dimension for sequence length
        X_batch = X_batch.unsqueeze(1)  # Assuming sequence length is 1 here
        
        # Forward pass
        outputs = model_test(X_batch)
        
        # Collect predictions and actual values
        predictions.append(outputs.squeeze().cpu().numpy())
        actuals.append(y_batch.cpu().numpy())

# Convert lists to numpy arrays for easier manipulation
predictions = np.concatenate(predictions, axis=0)
actuals = np.concatenate(actuals, axis=0)

# Print out the predicted and actual values
for i in range(len(predictions)):
    print(f"Predicted: {predictions[i]:.4f}, Actual: {actuals[i]:.4f}")
```
